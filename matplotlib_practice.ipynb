```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
#5.14 Skill Challenge
fig, ax  = plt.subplots()
ax.plot(x_axis, y_axis, marker = 'D', color = 'green', linewidth = 2, label = 'Boston')
ax.set_ylim(0, 45)
ax.set_xlabel('Date', weight="bold")
ax.set_ylabel('Fare ($)', weight="bold")
ax.set_title("Pyber Fare by Month", weight="bold")
ax.grid()
ax.legend()
```

```python
#5.1.7
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#skill Drill.
plt.scatter(y_axis,x_axis, c = "red", label = "Chicago")
plt.title("Pyber Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.legend()
plt.show()
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
#skill drill
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis, c = "skyblue", alpha = 0.8, edgecolor = "black", linewidth = 2, label = "Boston", s = [i * 5 for i in y_axis])
ax.set_title("Pyber Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0, 50)
ax.invert_yaxis()
ax.legend()
```

```python
#5.1.8
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, colors = colors, labels = x_axis, autopct = '%.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(y_axis, labels = x_axis)
plt.show()
```
